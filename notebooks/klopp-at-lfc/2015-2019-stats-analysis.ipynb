```python
import pandas as pd
import numpy as np
```

```python
# Jurgen Klopps tenure
# DateMatch Date (dd/mm/yy)
# HomeTeamHome Team
# AwayTeamAway Team
# FTHGFull Time Home Team Goals
# FTAGFull Time Away Team Goals
# FTRFull Time Result (H=Home Win, D=Draw, A=Away Win)
# HTHGHalf Time Home Team Goals
# HTAGHalf Time Away Team Goals
# HTRHalf Time Result (H=Home Win, D=Draw, A=Away Win)
# RefereeMatch Referee
# HSHome Team Shots
# ASAway Team Shots
# HSTHome Team Shots on Target
# ASTAway Team Shots on Target
# HCHome Team Corners
# ACAway Team Corners
# HFHome Team Fouls Committed
# AFAway Team Fouls Committed
# HYHome Team Yellow Cards
# AYAway Team Yellow Cards
# HRHome Team Red Cards
# ARAway Team Red Cards
```

```python
df = pd.read_csv("EPL.csv")
df = df[(df['HomeTeam']=='Liverpool') | (df['AwayTeam'] == 'Liverpool')]
df['Date'] = pd.to_datetime(df['Date']).astype(str)
df['Year'] = df['Date'].apply(lambda x: x.split('-')[0])
df['Month'] = df['Date'].apply(lambda x: x.split('-')[1]).astype(int)
df = df[(df['Year']>='2015')]
df = df.sort_values(by=['Year','Month']).reset_index(drop=True)
df = df[df['Date']>='2015-10-08']
home = df.copy(deep=True)
home = home[home['HomeTeam'] == 'Liverpool']
away = df.copy(deep=True)
away = away[away['AwayTeam']=='Liverpool']
home_goals = home.groupby(['HomeTeam','AwayTeam'])['FTHG'].sum().to_frame().sort_values(by='FTHG',ascending=False)
home_goals =home_goals.reset_index()
away_goals = away.groupby(['AwayTeam','HomeTeam'])['FTAG'].sum().to_frame().sort_values(by='FTAG',ascending=False)
away_goals = away_goals.reset_index()
combined_goals = pd.concat([home_goals,away_goals],axis=1)
```
